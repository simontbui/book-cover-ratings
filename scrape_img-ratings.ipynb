```python
import requests
from proxycrawl import CrawlingAPI
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote_plus
import json
import pprint
import ast
import pandas as pd
```

```python
token = "pIiVoNtQfnMuNI80pGbkKw"
```

```python
book_info = {"img": [], "rating": [], "num_rating": []}
```

```python
for i in range(1, 401):
    url = quote_plus(f"https://www.amazon.com/s?i=digital-text&rh=n%3A158571011&fs=true&page={i}")
    handler = urlopen(f"https://api.proxycrawl.com/?token={token}&scraper=amazon-serp&url={url}")
    serp = json.loads(handler.read().decode("utf-8"))
    prod_info = serp['body']['products']
    
    for ebook in prod_info:
        book_info["img"].append(ebook["image"])
        book_info["rating"].append(ebook["customerReview"])
        book_info["num_rating"].append(ebook["customerReviewCount"])
    
    
```

```python
pd.DataFrame(book_info)
```

```python
df = pd.DataFrame(book_info)
df = df.drop_duplicates()
df
```

```python
df.to_csv("ebook_info.csv", index=False)
```
